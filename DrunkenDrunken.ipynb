```
import matplotlib
matplotlib.use('Agg')
import random
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML

from celluloid import Camera as Camera
position= 10
N=100
#0,1 = atas,kanan 
#x=1 y=0
# 0,1,2,3= atas kanan bawah kiri
#x=1,3 y=0,3

range_step = 1



x2D=[]
y2D=[]
x4D=[]
y4D=[]
x2D.append(position)
y2D.append(position)
x4D.append(position)
y4D.append(position)
print(x2D,y2D,x4D,y4D)
i=1
d=[]
d.append(10)
e=[]
e.append(10)


while(i<N):
  long_x2=len(x2D)-1
  long_y2=len(y2D)-1
  long_x4=len(x4D)-1
  long_y4=len(y4D)-1
  dua = random.randint(0, 1)
  empat = random.randint(0, 3)
  e.append(empat)
  d.append(dua)
  if(dua==0):
    if(y2D[long_y2]+1 <=20):
      y2D.append(y2D[long_y2]+1)
      x2D.append(x2D[long_x2])
    else:
      y2D.append(random.randint(1, 19))
      x2D.append(random.randint(1, 19))
  elif(dua==1):
    if(x2D[long_x2]+1 <=20):
      x2D.append(x2D[long_x2]+1)
      y2D.append(y2D[long_y2])
    else:
      x2D.append(random.randint(1, 19))
      y2D.append(random.randint(1, 19))

  if(empat == 0):
     if(y4D[long_y4]+1 <=20):
       y4D.append(y4D[long_y4]+1)
       x4D.append(x4D[long_x4])
     else:
       y4D.append(random.randint(1, 19))
       x4D.append(random.randint(1, 19))
  elif(empat==1):
    if(x4D[long_x4]+1 <=20):
      x4D.append(x4D[long_x4]+1)
      y4D.append(y4D[long_y4])
    else:
      x4D.append(random.randint(1, 19))
      y4D.append(random.randint(1, 19))
  elif(empat==2):
    if(y4D[long_y4]-1 >=0):
      y4D.append(y4D[long_y4]-1)
      x4D.append(x4D[long_x4])
    else:
      y4D.append(random.randint(1, 19))
      x4D.append(random.randint(1, 19))
  elif(empat==3):
    if(x4D[long_x4]-1 >=0):
      x4D.append(x4D[long_x4]-1)
      y4D.append(y4D[long_y4])
    else:
      x4D.append(random.randint(1, 19))
      y4D.append(random.randint(1, 19))

  i=i+1


def dimensisatu(x2D,y2D):
    Camera1 = Camera(plt.figure())
    # array buat bikin plotan si animasi D1
    XD1= []
    YD1= []

    # plot animasi D1
    i=0
    while(i<N):  
        XD1.append(x2D[i])
        YD1.append(y2D[i])
        dimensi1 = [XD1, YD1]
        animasi_dimensi1= np.c_[dimensi1]
        plt.xlim(0, 20.2)
        plt.ylim(0, 20.2)
        plt.title("Drunken 1D animation plot")
        plt.scatter(*animasi_dimensi1, c="black", s=30, label="D1 - orang")
        plt.scatter(XD1[len(XD1)-1],YD1[len(YD1)-1], c='b')
        plt.scatter(XD1[0],YD1[0], c='r', marker='x')
        Camera1.snap()
        i=i+1

    #plot anim

    anim = Camera1.animate(interval=100)
    HTML(anim.to_html5_video())
    plt.grid(True, which="both")
    plt.show()


# array buat bikin plotan si animasi D2
def dimensidua(x4D,y4D):
    Camera2 = Camera(plt.figure())
    XD2= []
    YD2= []
    i=0
    while(i<N):
        XD2.append(x4D[i])
        YD2.append(y4D[i])
        dimensi2 = [XD2, YD2]
        animasi_dimensi2= np.c_[dimensi2]
    # plot animasi D4
        plt.xlim(0, 20.2)
        plt.ylim(0, 20.2)
        plt.title("Drunken 1D animation plot")
        plt.scatter(*animasi_dimensi2, c="black", s=30, label="D1 - orang")
        plt.scatter(XD2[len(XD2)-1],YD2[len(YD2)-1], c='b')
        plt.scatter(XD2[0],YD2[0], c='r', marker='x')
        Camera2.snap()
        i=i+1

    #plot anim
    anim = Camera2.animate(interval=100)
    HTML(anim.to_html5_video())
    plt.grid(True, which="both")
    plt.show() 




dimensisatu(x2D,y2D)
dimensidua(x4D,y4D)

# per plotan biasa

fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 20.2)
ax.set_ylim(0, 20.2)

ax.scatter(x2D,y2D, c='black')
ax.scatter(x2D[len(x2D)-1],y2D[len(y2D)-1], c='b')
ax.scatter(x2D[0],y2D[0], c='r', marker='x')
plt.title("A Drunkard’s Walk 1D")
plt.show()

fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 20.2)
ax.set_ylim(0, 20.2)
ax.scatter(x4D,y4D, c='black')
ax.scatter(x4D[len(x4D)-1],y4D[len(y4D)-1], c='b')
ax.scatter(x4D[0],y4D[0], c='r', marker='x')
plt.title("A Drunkard’s Walk 2D")
plt.show()

```
